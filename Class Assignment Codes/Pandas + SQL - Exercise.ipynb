```python
import pandas as pd
from pandasql import sqldf
import numpy as np
```

```python
# Load the AWProduct.csv file into a Pandas Dataframe. Name the Dataframe df
df = pd.read_csv('/Users/Name/Desktop/AWProduct.csv')
```

```python
# Using SQL: Display the first 10 rows of the Dataframe. 
# Only show the following columns: ProductKey, ProductAlternateKey, ProductSubcategoryKey, ProductName
sqlcode = """
    SELECT ProductKey, ProductAlternateKey, ProductSubcategoryKey, ProductName 
    FROM df 
    LIMIT 10"""
print(sqldf(sqlcode))
```

```python
# Using Pandas: Display the first 10 rows of the Dataframe. 
# Only show the following columns: ProductKey, ProductAlternateKey, ProductSubcategoryKey, ProductName
df[['ProductKey', 'ProductAlternateKey', 'ProductSubcategoryKey', 'ProductName']].head(10)
```

```python
# Using SQL: Display ProductName, DealerPrice only for those rows where the DealerPrice > 10. 
# Limit your output to only 10 rows
sqlcode1 = """
    SELECT ProductName, DealerPrice
    FROM df
    WHERE DealerPrice > 10
    LIMIT 10"""
print(sqldf(sqlcode1))
```

```python
# Using Pandas: Display ProductName, DealerPrice only for those rows where the DealerPrice > 10.
# Limit your output to only 10 rows
df.loc[df['DealerPrice'] > 10, ['ProductName', 'DealerPrice']].head(10)
```

```python
# Using SQL: Display ProductName, DealerPrice, Color only for those rows where the DealerPrice > 10 AND Color = Silver.
# Limit your output to only 10 rows
sqlcode2 = """
    SELECT ProductName, DealerPrice
    FROM df
    WHERE DealerPrice > 10 AND Color = 'Silver'
    LIMIT 10"""
print(sqldf(sqlcode2))
```

```python
# Using Pandas: Display ProductName, DealerPrice, Color only for those rows where the DealerPrice > 10 AND Color = Silver. 
# Limit your output to only 10 rows
df.loc[(df['DealerPrice'] > 10) & (df['Color'] == 'Silver'), ['ProductName', 'DealerPrice']].head(10)
```

```python
# Using SQL: Group by color. Find the ListPrice AVERAGE per color.
sqlcode3 = """
SELECT color, avg(ListPrice) as 'Average List Price'
FROM df
WHERE color != 'None'
GROUP BY color"""
print(sqldf(sqlcode3))
```

```python
# Using Pandas: Group by color. Find the ListPrice AVERAGE per color.
df.groupby(['Color'])['ListPrice'].mean()
```

```python
# Load the AWProductSubcategory.csv file into a Pandas Dataframe. Name the Dataframe df_s
df_s = pd.read_csv('/Users/jessietam/Desktop/AWProductSubcategory.csv')
```

```python
# Using SQL: Join the df and df_s Dataframes by the ProductSubcategoryKey column. 
# Display ProductAlternateKey, ProductName, ProductSubcategoryKey, ProductCategory, ProductSubcategoryName.
# Limit your output to only 10 rows
sqlcode4 = """
SELECT ProductAlternateKey, ProductName, df.ProductSubcategoryKey, ProductCategory, ProductSubcategoryName
FROM df, df_s
WHERE df.ProductSubcategoryKey = df_s.ProductSubcategoryKey
LIMIT 10"""
print(sqldf(sqlcode4))
```

```python
# Using Pandas: Join the df and df_s Dataframes by the ProductSubcategoryKey column.
# Display ProductAlternateKey, ProductName, ProductSubcategoryKey, ProductCategory, ProductSubcategoryName. 
# Limit your output to only 10 rows
merged = pd.merge(df, df_s, on='ProductSubcategoryKey')
merged[['ProductAlternateKey', 'ProductName', 'ProductSubcategoryKey', 'ProductCategory', 'ProductSubcategoryName']].head(10)
```
